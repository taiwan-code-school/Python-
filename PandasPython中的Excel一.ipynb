## 範例三：查看部分內容

可以透過下列方法查看目前資料的資訊：

* .shape   →回傳列數與欄數
* .describe() →回傳一些描述性統計，平均值、標準差、最小最大值..等
* .head() → 回傳前 5 筆資料內容，()內可自填想回傳從頭數來的「幾」筆資料
* .tail() → 回傳後 5 筆資料內容，()內可自填想回傳倒數「幾」筆資料
* .columns → 回傳所有欄位名稱
* .index → 回傳 index(索引值)有多少  
* .info() → 回傳資料內容的資訊

## Matplotlib 數據可視化的高手

```python
from matplotlib import pyplot as plt
```

matplotlib 就像是一個圖書館一樣，裡面有各種關於數據可視化的不同功能，


目前我們不需要把整個圖書館都搬回家，只需要他的其中一個功能「pyplot」，


所以我們就單獨約他出來就可以了>///<

那我們一樣幫他取個小名叫做 plt (比較親切嘛～XD)

各種圖示的功能：
* plt.plot() 折線圖
* plt.bar() 長條圖
* plt.scatter() 散布圖

()括號裡面放畫圖所需要的參數： x(橫坐標) 和 y(縱座標)，如果只放一個參數，預設是 y(縱座標) 的資料，x(橫坐標) 就會是預設的0,1,2,3…(看y有幾筆資料就會有幾筆)。

```python
plt.plot(df['地震個數'])
```

當我們把資訊都給 plt 之後，他會幫我們「默默的在腦海中」畫出來，因為他還不確定你是否要看圖了。

所以，我們必須還要明確的跟他說

plt.show()  ，意思就是：「幫我顯示一下圖嘿，甘溫啦～XD」 

→要注意show是一個功能，所以後面要加上()小括號才行哦!

plt.show()

### 小技巧

如果不想每一次都還要請他show()出來，也可以先在程式之前打上一個魔術指令

「%matplotlib inline」

這樣的話，就不用打plt.show()，Jupyter就會直接幫你顯示出來了。

```python
%matplotlib inline
plt.plot(df['地震個數'])
```
